```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Part 1 - Data Preprocessing
```

```python
# Importing the dataset
# this dataset is for a bank trying to determine churn rate. 
# It is a classification problem. Given the independent variables creditscore, gender, aget etc 
# we will predict the 'Exited' column (which indictes if the client will leave the bank or not)
dataset = pd.read_csv('ANN.csv')
print(dataset.head())
```

```python
print(dataset.shape)
```

```python
# independent variables are from 'CreditScore' to 'EstimatedSalary'. There are 11 independent variables. 
X = dataset.iloc[:, 3:13].values
# dependent variable (target) is the 'Exited' column
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data which are 'Geography' and 'Gender'
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling using StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Part 2 - Now let's make the ANN!
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential  # this is used to initialize the neural network
from keras.layers import Dense  # this is used to build the layers
from keras.layers import Dropout # dropout regularization to prevent overfitting
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# our neural network will have 2 hidden layers.
```

```python
# usually we select the average for the number of nodes for the input and output layers.
# in this example, we have 11 input nodes (independent variables) and 1 output node (depedent variable)
# so we will use (11+1)/2 = 6
```

```python
# overfitting is when you have a large difference between the accuracy (performance of the model) for training 
# and test datasets. You will have much higher accuracy on the training data set than the test dataset.

# "Dropout Regularization" is used to reduce overfitting if needed.
```

```python
# Adding the input layer and the first hidden layer
# this layer uses the 'rectifier' activation function
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# for Dropout try p values between 0.1 and 0.4
classifier.add(Dropout(p=0.1))
```

```python
# Adding the second hidden layer
# this payer uses the 'rectifier' activation function
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

classifier.add(Dropout(p=0.1))
```

```python
# Adding the output layer
# this layer uses the 'sigmoid' activation function since this is a classifier problem.
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
# there are several types of stochastic gradient descent algorithms. We will use 'adam'.
# binary_crossentropy is the logarithmic loss function.
# 'accuracy' criterion is used to optimize the weights.
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
# batch_size is the number of observations before weights are updated.
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
# accuracy is increasing as the algorithm runs. 
```

```python
# we got accuracy of about 83%
```

```python
# Part 3 - Making the predictions and evaluating the model
```

```python
# Predicting the Test set results.
# the test dataset was already scaled.

y_pred = classifier.predict(X_test)

# any output larger than 0.5 will be 'True', otherwise it will be 'False'. 
y_pred = (y_pred > 0.5)
```

```python
# Predicting a single new observation
"""Predict if the customer with the following information will leave the bank or not:
Geography: France
Credit Score: 600
Gender: Male
Age: 40
Tenure: 3
Balance: 60000
Number of Products: 2
Has Credit Card: Yes
Is Active Member: Yes
Estimated Salary: 50000"""
# double brackets create a 2D array with one row
# note that onehotencoding is applied manually, 
# that is, the first two values are the dummy variables for 'geography' and 'gender'
# sc.transform() performs the scaling of the input
new_prediction = classifier.predict(sc.transform(np.array([[0.0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])))
print(new_prediction)
```

```python
new_prediction = (new_prediction > 0.5)
print(new_prediction)
```

```python
# the customer will not leave the bank.
```

```python
# scoring the model
# Making the Confusion Matrix
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Part 4 - Evaluating, Improving and Tuning the ANN
```

```python
# Evaluating the ANN
# k-fold cross-validation belongs to sklearn  (not keras)
# but keras can provide a wrapper called KerasClassifier()
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

from keras.models import Sequential
from keras.layers import Dense

# KerasClassifier() requires a function as parameter
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

# KerasClassifier() is the wrapper
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)

# to measure the model performance, we will use k-fold cross-validation
# n_jobs = -1 means, it will use all CPUs
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10, n_jobs = -1)

mean = accuracies.mean()
print(mean)
variance = accuracies.std()
print(variance)
```

```python
# Improving the ANN
```

```python
# Tuning the ANN by using GridSearchCV
# GridSearchCV belongs to sklearn  (not keras)
# but keras can provide a wrapper called KerasClassifier()
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense

# KerasClassifier() requires a function as parameter
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

# KerasClassifier() is the wrapper
classifier = KerasClassifier(build_fn = build_classifier)

parameters = {'batch_size': [25, 32],
              'epochs': [100, 500],
              'optimizer': ['adam', 'rmsprop']}

grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10)

grid_search = grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
print(best_parameters)
best_accuracy = grid_search.best_score_
print(best_accuracy)
```
