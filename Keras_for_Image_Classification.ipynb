# Keras / ResNet50 for Image Classification

### In this example, we will be using ResNet50 for classifying an image

```python
# load the required libraries
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_PATH = 'keras_for_image_classification/sample_images/bicycle.jpg'
```

```python
# display the image
img = image.load_img(IMG_PATH, target_size=(224, 224))
plt.imshow(img)
plt.show()
```

```python
# define the model to use
model = tf.keras.applications.resnet50.ResNet50()
```

```python
# here is a simple function that loads the image and uses ResNet50

def predict(img_path):
    # load the image, resize to 224 x 224
    img = image.load_img(img_path, target_size=(224, 224))
    # define the model
    model = tf.keras.applications.resnet50.ResNet50()
    img_array = image.img_to_array(img)
    img_batch = np.expand_dims(img_array, axis=0)
    # initiatiate pre-processing
    img_preprocessed = preprocess_input(img_batch)
    # run the model
    prediction = model.predict(img_preprocessed)
    # return only the top 3 predictions
    print(decode_predictions(prediction, top=3)[0])
```

```python
predict(IMG_PATH)
```
